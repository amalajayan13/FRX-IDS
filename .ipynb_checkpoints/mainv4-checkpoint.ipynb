# FRX-IDS Clean Fixed Notebook — ToN-IoT Primary Evaluation + UNSW-NB15 External Validation

This version was rebuilt to run cleanly in Google Colab with Google Drive as the default storage location. It fixes the previously observed `log`/`X_train` dependency errors by using consistent progressive logging, robust feature-variable recovery, and safer explainability logic. Each step saves outputs to Google Drive under `MyDrive/Outputs/FRX_IDS_UNSW_VALIDATION`.


## Cell 1 — Environment, Google Drive, Paths, and Progressive Logging

```python
# ============================================================
# CELL 1 — ENVIRONMENT, GOOGLE DRIVE, PATHS, AND PROGRESSIVE LOGGING
# ============================================================

import os
import sys
import json
import time
import math
import uuid
import random
import hashlib
import warnings
import platform
import datetime
from pathlib import Path

import numpy as np
import pandas as pd

warnings.filterwarnings("ignore")

# -----------------------------
# Reproducibility
# -----------------------------
RANDOM_SEED = 42
random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)

# # -----------------------------
# # Storage mode
# # -----------------------------
# USE_GOOGLE_DRIVE = True   # DEFAULT, as requested
# USE_LOCAL_FILES = False   # Optional fallback only if set manually

# if USE_GOOGLE_DRIVE:
#     try:
#         from google.colab import drive
#         print("=" * 80)
#         print("MOUNTING GOOGLE DRIVE")
#         print("=" * 80)
#         drive.mount('/content/drive')
#         DRIVE_BASE = Path('/content/drive/MyDrive')
#     except Exception as e:
#         raise RuntimeError(f"Google Drive mode is enabled, but Drive mounting failed: {e}")
# else:
#     DRIVE_BASE = Path.cwd()

# -----------------------------
# Storage mode
# -----------------------------
from pathlib import Path
import os

# Automatically detect whether running inside Google Colab
try:
    import google.colab
    IN_COLAB = True
except ImportError:
    IN_COLAB = False

# Use Google Drive only when running in Colab
USE_GOOGLE_DRIVE = IN_COLAB
USE_LOCAL_FILES = not IN_COLAB

if USE_GOOGLE_DRIVE:
    try:
        from google.colab import drive
        print("=" * 80)
        print("MOUNTING GOOGLE DRIVE")
        print("=" * 80)
        drive.mount('/content/drive')
        DRIVE_BASE = Path('/content/drive/MyDrive')
    except Exception as e:
        raise RuntimeError(f"Google Drive mode is enabled, but Drive mounting failed: {e}")
else:
    DRIVE_BASE = Path.cwd()
    print("=" * 80)
    print("RUNNING IN LOCAL MODE")
    print("=" * 80)
    print("Working directory:", DRIVE_BASE)


# -----------------------------
# Dataset paths
# -----------------------------
if USE_GOOGLE_DRIVE:
    DATASET_BASE = DRIVE_BASE / 'Datasets' / 'Telecom'
    BASE_OUTPUT_DIR = DRIVE_BASE / 'Outputs' / 'FRX_IDS_UNSW_VALIDATION'
else:

    CODE_DIR = Path.cwd()
    DATASET_BASE = CODE_DIR.parent

    # Outputs will be created inside the working/code directory
    BASE_OUTPUT_DIR = CODE_DIR / 'Outputs' / 'FRX_IDS_UNSW_VALIDATION'

    
# Robust dataset resolver: supports common folder spellings used in Drive.
def resolve_dataset_path(candidates, recursive_pattern=None):
    for candidate in candidates:
        candidate = Path(candidate)
        if candidate.exists():
            return candidate
    if recursive_pattern is not None and DATASET_BASE.exists():
        matches = sorted(DATASET_BASE.rglob(recursive_pattern))
        if matches:
            return matches[0]
    return Path(candidates[0])

TON_IOT_PATH = resolve_dataset_path([
    DATASET_BASE / 'TON_IoT' / 'ton-iot-train_test_network.csv',
    DATASET_BASE / 'ToN_IoT' / 'ton-iot-train_test_network.csv',
    DATASET_BASE / 'ToN-IoT' / 'ton-iot-train_test_network.csv',
    DATASET_BASE / 'ton-iot-train_test_network.csv',
], recursive_pattern='*ton*iot*network*.csv')

UNSW_TRAIN_PATH = resolve_dataset_path([
    DATASET_BASE / 'UNSW_NB15' / 'UNSW_NB15_training-set.csv',
    DATASET_BASE / 'UNSW-NB15' / 'UNSW_NB15_training-set.csv',
    DATASET_BASE / 'UNSW_NB15_training-set.csv',
    DATASET_BASE / 'UNSW-NB15_training-set.csv',
], recursive_pattern='*UNSW*training*.csv')

UNSW_TEST_PATH = resolve_dataset_path([
    DATASET_BASE / 'UNSW_NB15' / 'UNSW_NB15_testing-set.csv',
    DATASET_BASE / 'UNSW-NB15' / 'UNSW_NB15_testing-set.csv',
    DATASET_BASE / 'UNSW_NB15_testing-set.csv',
    DATASET_BASE / 'UNSW-NB15_testing-set.csv',
], recursive_pattern='*UNSW*testing*.csv')

FIG_DIR = BASE_OUTPUT_DIR / 'Figures'
TABLE_DIR = BASE_OUTPUT_DIR / 'Tables'
MODEL_DIR = BASE_OUTPUT_DIR / 'Models'
OUTPUT_DIR = BASE_OUTPUT_DIR / 'Outputs'
LOG_DIR = BASE_OUTPUT_DIR / 'Logs'
NOTEBOOK_DIR = BASE_OUTPUT_DIR / 'Notebook'

for d in [BASE_OUTPUT_DIR, FIG_DIR, TABLE_DIR, MODEL_DIR, OUTPUT_DIR, LOG_DIR, NOTEBOOK_DIR]:
    d.mkdir(parents=True, exist_ok=True)

OUTPUTS_SUMMARY_PATH = OUTPUT_DIR / 'outputs_summary.txt'
ARTIFACT_REGISTRY_PATH = OUTPUT_DIR / 'artifact_registry.csv'
RUN_MANIFEST_PATH = OUTPUT_DIR / 'run_manifest.json'

ARTIFACTS = []

# -----------------------------
# Progressive logging helpers
# -----------------------------
def utc_now():
    return datetime.datetime.utcnow().replace(microsecond=0).isoformat() + 'Z'

def write_text(path, text, mode='w'):
    path = Path(path)
    path.parent.mkdir(parents=True, exist_ok=True)
    with open(path, mode, encoding='utf-8') as f:
        f.write(text)

def append_summary(message):
    stamp = utc_now()
    line = f"[{stamp}] {message}\n"
    write_text(OUTPUTS_SUMMARY_PATH, line, mode='a')
    print(line.rstrip())

# Backward-compatible logger alias.
# This prevents failures in older cells that may call log(...).
log = append_summary

def register_artifact(path, artifact_type, description):
    path = Path(path)
    size_mb = path.stat().st_size / (1024 * 1024) if path.exists() else 0.0
    row = {
        'timestamp_utc': utc_now(),
        'artifact_type': artifact_type,
        'description': description,
        'path': str(path),
        'filename': path.name,
        'exists': path.exists(),
        'size_mb': round(size_mb, 6)
    }
    ARTIFACTS.append(row)
    pd.DataFrame(ARTIFACTS).to_csv(ARTIFACT_REGISTRY_PATH, index=False)
    append_summary(f"ARTIFACT SAVED | {artifact_type} | {description} | {path}")

def save_json(obj, path, description):
    path = Path(path)
    path.parent.mkdir(parents=True, exist_ok=True)
    with open(path, 'w', encoding='utf-8') as f:
        json.dump(obj, f, indent=2, default=str)
    register_artifact(path, 'json', description)

def save_table(df, path, description):
    path = Path(path)
    path.parent.mkdir(parents=True, exist_ok=True)
    df.to_csv(path, index=False)
    register_artifact(path, 'table_csv', description)

def save_figure(path, description):
    path = Path(path)
    register_artifact(path, 'figure_png', description)

def refresh_outputs_summary(final=False):
    lines = []
    lines.append('=' * 80)
    lines.append('FRX-IDS CLEAN PIPELINE OUTPUTS SUMMARY')
    lines.append('=' * 80)
    lines.append(f'Last refreshed UTC : {utc_now()}')
    lines.append(f'Final refresh      : {final}')
    lines.append('')
    lines.append('DIRECTORIES')
    lines.append('-' * 80)
    for name, path in [
        ('BASE_OUTPUT_DIR', BASE_OUTPUT_DIR),
        ('FIG_DIR', FIG_DIR),
        ('TABLE_DIR', TABLE_DIR),
        ('MODEL_DIR', MODEL_DIR),
        ('OUTPUT_DIR', OUTPUT_DIR),
        ('LOG_DIR', LOG_DIR),
        ('NOTEBOOK_DIR', NOTEBOOK_DIR),
    ]:
        lines.append(f'{name}: {path}')
    lines.append('')
    lines.append('DATASETS')
    lines.append('-' * 80)
    for name, path in [('ToN-IoT', TON_IOT_PATH), ('UNSW train', UNSW_TRAIN_PATH), ('UNSW test', UNSW_TEST_PATH)]:
        exists = path.exists()
        size = path.stat().st_size / (1024 * 1024) if exists else 0.0
        lines.append(f'{name}: {path} | exists={exists} | size_mb={size:.4f}')
    lines.append('')
    lines.append('REGISTERED ARTIFACTS')
    lines.append('-' * 80)
    if ARTIFACTS:
        for a in ARTIFACTS:
            lines.append(f"{a['timestamp_utc']} | {a['artifact_type']} | {a['description']} | {a['path']} | size_mb={a['size_mb']}")
    else:
        lines.append('No artifacts registered yet.')
    lines.append('')
    lines.append('DIRECTORY FILE COUNTS')
    lines.append('-' * 80)
    for path in [FIG_DIR, TABLE_DIR, MODEL_DIR, OUTPUT_DIR, LOG_DIR, NOTEBOOK_DIR]:
        files = [p for p in path.glob('*') if p.is_file()]
        lines.append(f'{path}: {len(files)} files')
        for p in sorted(files)[:50]:
            lines.append(f'  - {p.name} ({p.stat().st_size / (1024 * 1024):.4f} MB)')
    write_text(OUTPUTS_SUMMARY_PATH, '\n'.join(lines) + '\n', mode='w')

# Initialize outputs_summary.txt immediately
refresh_outputs_summary(final=False)
append_summary('Cell 1 completed: environment, paths, and progressive logging initialized.')

run_manifest = {
    'created_utc': utc_now(),
    'python': sys.version,
    'platform': platform.platform(),
    'numpy': np.__version__,
    'pandas': pd.__version__,
    'random_seed': RANDOM_SEED,
    'use_google_drive': USE_GOOGLE_DRIVE,
    'ton_iot_path': str(TON_IOT_PATH),
    'unsw_train_path': str(UNSW_TRAIN_PATH),
    'unsw_test_path': str(UNSW_TEST_PATH),
    'base_output_dir': str(BASE_OUTPUT_DIR),
    'outputs_summary_path': str(OUTPUTS_SUMMARY_PATH)
}
save_json(run_manifest, RUN_MANIFEST_PATH, 'Run manifest with paths and environment')

print('\n[Configured paths]')
print('ToN-IoT        :', TON_IOT_PATH)
print('UNSW train     :', UNSW_TRAIN_PATH)
print('UNSW test      :', UNSW_TEST_PATH)
print('Outputs summary:', OUTPUTS_SUMMARY_PATH)
```

## Cell 2 — Dataset Loading and Initial Inspection

```python
# ============================================================
# CELL 2 — DATASET LOADING AND INITIAL INSPECTION
# ============================================================

print("=" * 80)
print("CELL 2 — DATASET LOADING AND INITIAL INSPECTION")
print("=" * 80)

for name, path in {'ToN-IoT': TON_IOT_PATH, 'UNSW train': UNSW_TRAIN_PATH, 'UNSW test': UNSW_TEST_PATH}.items():
    if not path.exists():
        raise FileNotFoundError(f"{name} dataset not found: {path}")
    append_summary(f"Dataset found: {name} | {path} | size_mb={path.stat().st_size/(1024*1024):.4f}")

def safe_read_csv(path):
    t0 = time.perf_counter()
    try:
        df = pd.read_csv(path, low_memory=False)
    except UnicodeDecodeError:
        df = pd.read_csv(path, low_memory=False, encoding='latin1')
    load_time = time.perf_counter() - t0
    return df, load_time

ton_df, ton_load_time = safe_read_csv(TON_IOT_PATH)
unsw_train_df, unsw_train_load_time = safe_read_csv(UNSW_TRAIN_PATH)
unsw_test_df, unsw_test_load_time = safe_read_csv(UNSW_TEST_PATH)

print('\n[Loaded shapes]')
print('ToN-IoT    :', ton_df.shape)
print('UNSW train :', unsw_train_df.shape)
print('UNSW test  :', unsw_test_df.shape)

summary_rows = []
for name, df, path, load_time in [
    ('ToN-IoT', ton_df, TON_IOT_PATH, ton_load_time),
    ('UNSW-NB15 train', unsw_train_df, UNSW_TRAIN_PATH, unsw_train_load_time),
    ('UNSW-NB15 test', unsw_test_df, UNSW_TEST_PATH, unsw_test_load_time),
]:
    summary_rows.append({
        'dataset': name,
        'path': str(path),
        'rows': int(df.shape[0]),
        'columns': int(df.shape[1]),
        'load_time_seconds': round(load_time, 4),
        'memory_mb': round(df.memory_usage(deep=True).sum()/(1024*1024), 4),
        'duplicate_rows': int(df.duplicated().sum()),
        'missing_values': int(df.isna().sum().sum())
    })

dataset_summary_df = pd.DataFrame(summary_rows)
save_table(dataset_summary_df, TABLE_DIR / 'table_02_dataset_loading_summary.csv', 'Dataset loading summary')
save_json({r['dataset']: r for r in summary_rows}, OUTPUT_DIR / 'cell_02_dataset_loading_summary.json', 'Dataset loading summary JSON')
refresh_outputs_summary(final=False)
append_summary('Cell 2 completed: datasets loaded and inspection artifacts saved.')
display(dataset_summary_df)
```

## Cell 3 — Primary ToN-IoT Multiclass Preparation

```python
# ============================================================
# CELL 3 — PRIMARY ToN-IoT MULTICLASS PREPARATION
# ============================================================

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

print("=" * 80)
print("CELL 3 — PRIMARY ToN-IoT MULTICLASS PREPARATION")
print("=" * 80)

assert 'type' in ton_df.columns, "ToN-IoT column 'type' not found."

df_ton = ton_df.copy()
df_ton = df_ton.replace([np.inf, -np.inf], np.nan)
df_ton = df_ton.dropna(subset=['type']).copy()

# Fill missing values before splitting
for c in df_ton.columns:
    if df_ton[c].isna().any():
        if pd.api.types.is_numeric_dtype(df_ton[c]):
            df_ton[c] = df_ton[c].fillna(df_ton[c].median())
        else:
            mode = df_ton[c].mode(dropna=True)
            df_ton[c] = df_ton[c].fillna(mode.iloc[0] if len(mode) else 'unknown')

label_encoder_ton = LabelEncoder()
y_ton = label_encoder_ton.fit_transform(df_ton['type'].astype(str))
CLASS_NAMES_TON = label_encoder_ton.classes_
CLASS_NAMES = CLASS_NAMES_TON  # backward-compatible alias

leakage_cols = [c for c in ['type', 'label'] if c in df_ton.columns]
X_ton = df_ton.drop(columns=leakage_cols)

X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    X_ton, y_ton, test_size=0.20, random_state=RANDOM_SEED, stratify=y_ton
)
X_train_raw, X_val_raw, y_train, y_val = train_test_split(
    X_train_raw, y_train, test_size=0.20, random_state=RANDOM_SEED, stratify=y_train
)

print('X_train_raw:', X_train_raw.shape)
print('X_val_raw  :', X_val_raw.shape)
print('X_test_raw :', X_test_raw.shape)
print('Classes    :', list(CLASS_NAMES_TON))

class_distribution_df = pd.DataFrame({
    'class_id': np.arange(len(CLASS_NAMES_TON)),
    'class_name': CLASS_NAMES_TON,
    'count': [(y_ton == i).sum() for i in range(len(CLASS_NAMES_TON))],
    'percent': [round(100 * (y_ton == i).mean(), 4) for i in range(len(CLASS_NAMES_TON))]
})
save_table(class_distribution_df, TABLE_DIR / 'table_03_ton_class_distribution.csv', 'ToN-IoT multiclass distribution')

prep_summary = {
    'dataset': 'ToN-IoT',
    'target_column': 'type',
    'dropped_columns': leakage_cols,
    'train_shape': X_train_raw.shape,
    'validation_shape': X_val_raw.shape,
    'test_shape': X_test_raw.shape,
    'classes': CLASS_NAMES_TON.tolist()
}
save_json(prep_summary, OUTPUT_DIR / 'cell_03_ton_preparation_summary.json', 'ToN-IoT preparation summary')
refresh_outputs_summary(final=False)
append_summary('Cell 3 completed: ToN-IoT multiclass target prepared and split.')
display(class_distribution_df)
```

## Cell 4 — Feature Encoding and Numeric Matrix Construction

```python
# ============================================================
# CELL 4 — FEATURE ENCODING AND NUMERIC MATRIX CONSTRUCTION
# ============================================================

from sklearn.preprocessing import LabelEncoder

print("=" * 80)
print("CELL 4 — FEATURE ENCODING AND NUMERIC MATRIX CONSTRUCTION")
print("=" * 80)

X_train = X_train_raw.copy()
X_val = X_val_raw.copy()
X_test = X_test_raw.copy()

categorical_cols = X_train.select_dtypes(exclude=[np.number]).columns.tolist()
numeric_cols = X_train.select_dtypes(include=[np.number]).columns.tolist()
feature_encoders = {}

for c in categorical_cols:
    le = LabelEncoder()
    X_train[c] = X_train[c].astype(str)
    X_val[c] = X_val[c].astype(str)
    X_test[c] = X_test[c].astype(str)
    le.fit(X_train[c])
    classes = set(le.classes_)
    X_val[c] = X_val[c].apply(lambda x: x if x in classes else 'UNKNOWN')
    X_test[c] = X_test[c].apply(lambda x: x if x in classes else 'UNKNOWN')
    if 'UNKNOWN' not in le.classes_:
        le.classes_ = np.append(le.classes_, 'UNKNOWN')
    X_train[c] = le.transform(X_train[c])
    X_val[c] = le.transform(X_val[c])
    X_test[c] = le.transform(X_test[c])
    feature_encoders[c] = le

for X_part in [X_train, X_val, X_test]:
    X_part.replace([np.inf, -np.inf], np.nan, inplace=True)
    for c in X_part.columns:
        X_part[c] = pd.to_numeric(X_part[c], errors='coerce')

train_medians = X_train.median(numeric_only=True).fillna(0)
X_train = X_train.fillna(train_medians).fillna(0).astype(np.float32)
X_val = X_val.fillna(train_medians).fillna(0).astype(np.float32)
X_test = X_test.fillna(train_medians).fillna(0).astype(np.float32)

assert np.isfinite(X_train.values).all()
assert np.isfinite(X_val.values).all()
assert np.isfinite(X_test.values).all()

feature_metadata_df = pd.DataFrame({
    'feature': X_train.columns,
    'dtype_after_encoding': X_train.dtypes.astype(str).values,
    'was_categorical': [c in categorical_cols for c in X_train.columns]
})
save_table(feature_metadata_df, TABLE_DIR / 'table_04_feature_metadata.csv', 'Feature metadata after encoding')

save_json({
    'n_numeric_original': len(numeric_cols),
    'n_categorical_original': len(categorical_cols),
    'n_features_final': int(X_train.shape[1]),
    'train_shape': X_train.shape,
    'validation_shape': X_val.shape,
    'test_shape': X_test.shape,
    'categorical_columns': categorical_cols
}, OUTPUT_DIR / 'cell_04_feature_encoding_summary.json', 'Feature encoding summary')

refresh_outputs_summary(final=False)
append_summary('Cell 4 completed: features encoded and numeric matrices created.')
display(feature_metadata_df.head(20))
```

## Cell 5 — Model Training and Selection on ToN-IoT

```python
# ============================================================
# CELL 5 — MODEL TRAINING AND SELECTION ON ToN-IoT
# ============================================================

import joblib
from sklearn.metrics import accuracy_score, balanced_accuracy_score, precision_score, recall_score, f1_score
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, HistGradientBoostingClassifier

print("=" * 80)
print("CELL 5 — MODEL TRAINING AND SELECTION ON ToN-IoT")
print("=" * 80)

models = {}
try:
    from lightgbm import LGBMClassifier
    models['LightGBM'] = LGBMClassifier(
        objective='multiclass', num_class=len(CLASS_NAMES_TON), n_estimators=350,
        learning_rate=0.05, num_leaves=63, subsample=0.9, colsample_bytree=0.9,
        random_state=RANDOM_SEED, n_jobs=-1, verbose=-1
    )
except Exception as e:
    print('[WARN] LightGBM unavailable:', e)

try:
    from xgboost import XGBClassifier
    models['XGBoost'] = XGBClassifier(
        objective='multi:softprob', eval_metric='mlogloss', n_estimators=350,
        learning_rate=0.05, max_depth=7, subsample=0.9, colsample_bytree=0.9,
        random_state=RANDOM_SEED, n_jobs=-1, tree_method='hist'
    )
except Exception as e:
    print('[WARN] XGBoost unavailable:', e)

models['ExtraTrees'] = ExtraTreesClassifier(n_estimators=350, random_state=RANDOM_SEED, n_jobs=-1, class_weight='balanced')
models['RandomForest'] = RandomForestClassifier(n_estimators=300, random_state=RANDOM_SEED, n_jobs=-1, class_weight='balanced_subsample')
models['HistGradientBoosting'] = HistGradientBoostingClassifier(max_iter=250, learning_rate=0.06, random_state=RANDOM_SEED)

trained_models = {}
rows = []
for name, model in models.items():
    append_summary(f'Cell 5 training started: {name}')
    t0 = time.perf_counter()
    try:
        model.fit(X_train, y_train)
        train_time = time.perf_counter() - t0
        pred_train = model.predict(X_train)
        pred_val = model.predict(X_val)
        row = {
            'model': name,
            'train_accuracy': accuracy_score(y_train, pred_train),
            'val_accuracy': accuracy_score(y_val, pred_val),
            'val_balanced_accuracy': balanced_accuracy_score(y_val, pred_val),
            'val_precision_macro': precision_score(y_val, pred_val, average='macro', zero_division=0),
            'val_recall_macro': recall_score(y_val, pred_val, average='macro', zero_division=0),
            'val_f1_macro': f1_score(y_val, pred_val, average='macro', zero_division=0),
            'val_f1_weighted': f1_score(y_val, pred_val, average='weighted', zero_division=0),
            'train_time_seconds': train_time,
        }
        row['overfit_gap_train_minus_val'] = row['train_accuracy'] - row['val_accuracy']
        rows.append(row)
        trained_models[name] = model
        append_summary(f"Cell 5 model completed: {name} | val_macro_f1={row['val_f1_macro']:.6f}")
    except Exception as e:
        append_summary(f'Cell 5 model failed: {name} | {e}')

model_selection_df = pd.DataFrame(rows).sort_values(
    by=['val_f1_macro', 'val_balanced_accuracy', 'val_accuracy'], ascending=False
).reset_index(drop=True)
assert len(model_selection_df) > 0, 'No model trained successfully.'

best_model_name = model_selection_df.loc[0, 'model']
best_model = trained_models[best_model_name]
final_best_model_name = best_model_name
final_best_model = best_model

save_table(model_selection_df, TABLE_DIR / 'table_05_ton_model_selection.csv', 'ToN-IoT model selection results')
model_path = MODEL_DIR / f'model_05_final_ton_{best_model_name}.joblib'
joblib.dump(final_best_model, model_path)
register_artifact(model_path, 'model_joblib', f'Final ToN-IoT selected model: {best_model_name}')
save_json({'best_model_name': best_model_name, 'selection_table': str(TABLE_DIR / 'table_05_ton_model_selection.csv')}, OUTPUT_DIR / 'cell_05_model_selection_summary.json', 'Model selection summary')
refresh_outputs_summary(final=False)
append_summary(f'Cell 5 completed: best model selected = {best_model_name}.')
display(model_selection_df)
```

## Cell 6 — ToN-IoT Final Test Evaluation

```python
# ============================================================
# CELL 6 — ToN-IoT FINAL TEST EVALUATION
# ============================================================

from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
import matplotlib.pyplot as plt

print("=" * 80)
print("CELL 6 — ToN-IoT FINAL TEST EVALUATION")
print("=" * 80)

y_pred_ton = final_best_model.predict(X_test)

if hasattr(final_best_model, 'predict_proba'):
    y_proba_ton = final_best_model.predict_proba(X_test)
else:
    y_proba_ton = None

final_test_metrics = {
    'dataset': 'ToN-IoT',
    'model': final_best_model_name,
    'accuracy': accuracy_score(y_test, y_pred_ton),
    'balanced_accuracy': balanced_accuracy_score(y_test, y_pred_ton),
    'precision_macro': precision_score(y_test, y_pred_ton, average='macro', zero_division=0),
    'recall_macro': recall_score(y_test, y_pred_ton, average='macro', zero_division=0),
    'f1_macro': f1_score(y_test, y_pred_ton, average='macro', zero_division=0),
    'f1_weighted': f1_score(y_test, y_pred_ton, average='weighted', zero_division=0)
}
if y_proba_ton is not None:
    try:
        final_test_metrics['roc_auc_ovr_weighted'] = roc_auc_score(y_test, y_proba_ton, multi_class='ovr', average='weighted')
    except Exception:
        final_test_metrics['roc_auc_ovr_weighted'] = None

metrics_df = pd.DataFrame([final_test_metrics])
save_table(metrics_df, TABLE_DIR / 'table_06_ton_final_test_metrics.csv', 'ToN-IoT final test metrics')

report_df = pd.DataFrame(classification_report(y_test, y_pred_ton, target_names=CLASS_NAMES_TON, output_dict=True, zero_division=0)).T
save_table(report_df.reset_index().rename(columns={'index': 'class'}), TABLE_DIR / 'table_06_ton_classification_report.csv', 'ToN-IoT classification report')

cm = confusion_matrix(y_test, y_pred_ton)
cm_df = pd.DataFrame(cm, index=CLASS_NAMES_TON, columns=CLASS_NAMES_TON)
save_table(cm_df.reset_index().rename(columns={'index': 'true_class'}), TABLE_DIR / 'table_06_ton_confusion_matrix.csv', 'ToN-IoT confusion matrix')

plt.figure(figsize=(9, 7))
plt.imshow(cm, interpolation='nearest')
plt.title(f'ToN-IoT Confusion Matrix — {final_best_model_name}')
plt.xlabel('Predicted class')
plt.ylabel('True class')
plt.xticks(range(len(CLASS_NAMES_TON)), CLASS_NAMES_TON, rotation=45, ha='right')
plt.yticks(range(len(CLASS_NAMES_TON)), CLASS_NAMES_TON)
plt.colorbar()
plt.tight_layout()
fig_path = FIG_DIR / 'fig_06_ton_confusion_matrix.png'
plt.savefig(fig_path, dpi=300, bbox_inches='tight')
plt.show()
save_figure(fig_path, 'ToN-IoT confusion matrix figure')

save_json(final_test_metrics, OUTPUT_DIR / 'cell_06_ton_final_test_metrics.json', 'ToN-IoT final test metrics JSON')
refresh_outputs_summary(final=False)
append_summary('Cell 6 completed: ToN-IoT final test evaluation saved.')
display(metrics_df)
```

## Cell 7 — Learning Curves and Statistical Testing

```python
# ============================================================
# CELL 7 — FAST LEARNING CURVES AND STATISTICAL TESTING
# ============================================================

from sklearn.base import clone
from sklearn.model_selection import StratifiedKFold, learning_curve, cross_val_predict
from sklearn.metrics import make_scorer, f1_score, accuracy_score, precision_score, recall_score
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import time

print("=" * 80)
print("CELL 7 — FAST LEARNING CURVES AND STATISTICAL TESTING")
print("=" * 80)

# ============================================================
# Safety fallback utilities
# ============================================================

if 'append_summary' not in globals():
    def append_summary(message):
        print(message)

if 'save_table' not in globals():
    def save_table(df, path, description='table'):
        path.parent.mkdir(parents=True, exist_ok=True)
        df.to_csv(path, index=False)
        print(f"Saved {description}: {path}")

if 'save_figure' not in globals():
    def save_figure(path, description='figure'):
        print(f"Saved {description}: {path}")

# ============================================================
# Validation checks
# ============================================================

required_vars = [
    'X_train', 'X_val', 'y_train', 'y_val',
    'final_best_model', 'final_best_model_name',
    'FIG_DIR', 'TABLE_DIR', 'RANDOM_SEED'
]

missing = [v for v in required_vars if v not in globals()]
if missing:
    raise NameError(
        f"Missing required variables for Cell 7: {missing}\n"
        "Please execute Cells 1–6 first."
    )

# ============================================================
# Prepare combined train-validation set
# ============================================================

print("Preparing combined training-validation dataset...")

if isinstance(X_train, pd.DataFrame):
    X_trainval = pd.concat([X_train, X_val], axis=0).reset_index(drop=True)
    FEATURE_NAMES = X_trainval.columns.tolist()
else:
    X_trainval = np.concatenate([np.asarray(X_train), np.asarray(X_val)], axis=0)
    FEATURE_NAMES = [f'feature_{i}' for i in range(X_trainval.shape[1])]

if isinstance(y_train, pd.Series):
    y_trainval = pd.concat([y_train, y_val], axis=0).values
else:
    y_trainval = np.concatenate([np.asarray(y_train), np.asarray(y_val)])

print(f"Combined dataset shape: {X_trainval.shape}")
append_summary('Cell 7 fast learning curve computation started.')

# ============================================================
# Fast learning curve
# ============================================================

cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=RANDOM_SEED)
train_sizes = np.linspace(0.3, 1.0, 4)
scorer = make_scorer(f1_score, average='macro', zero_division=0)

start_time = time.time()
train_sizes_abs, train_scores, val_scores = learning_curve(
    estimator=clone(final_best_model),
    X=X_trainval,
    y=y_trainval,
    train_sizes=train_sizes,
    cv=cv,
    scoring=scorer,
    n_jobs=-1,
    shuffle=True,
    random_state=RANDOM_SEED
)
elapsed = time.time() - start_time
print(f"Learning curve computation completed in {elapsed:.2f} seconds")

learning_curve_df = pd.DataFrame({
    'train_size': train_sizes_abs,
    'train_f1_macro_mean': train_scores.mean(axis=1),
    'train_f1_macro_std': train_scores.std(axis=1),
    'validation_f1_macro_mean': val_scores.mean(axis=1),
    'validation_f1_macro_std': val_scores.std(axis=1)
})

print("\nLearning Curve Results")
display(learning_curve_df)

save_table(
    learning_curve_df,
    TABLE_DIR / 'table_07_fast_learning_curve.csv',
    'Fast learning curve table'
)

plt.figure(figsize=(8, 5))
plt.plot(
    learning_curve_df['train_size'],
    learning_curve_df['train_f1_macro_mean'],
    marker='o',
    linewidth=2,
    label='Training Macro-F1'
)
plt.plot(
    learning_curve_df['train_size'],
    learning_curve_df['validation_f1_macro_mean'],
    marker='o',
    linewidth=2,
    label='Validation Macro-F1'
)
plt.fill_between(
    learning_curve_df['train_size'],
    learning_curve_df['train_f1_macro_mean'] - learning_curve_df['train_f1_macro_std'],
    learning_curve_df['train_f1_macro_mean'] + learning_curve_df['train_f1_macro_std'],
    alpha=0.2
)
plt.fill_between(
    learning_curve_df['train_size'],
    learning_curve_df['validation_f1_macro_mean'] - learning_curve_df['validation_f1_macro_std'],
    learning_curve_df['validation_f1_macro_mean'] + learning_curve_df['validation_f1_macro_std'],
    alpha=0.2
)
plt.xlabel('Training Examples')
plt.ylabel('Macro-F1 Score')
plt.title(f'Fast Learning Curve — {final_best_model_name}')
plt.legend()
plt.grid(True)
plt.tight_layout()
fig_lc_path = FIG_DIR / 'fig_07_fast_learning_curve.png'
plt.savefig(fig_lc_path, dpi=300, bbox_inches='tight')
save_figure(fig_lc_path, 'Fast learning curve figure')
plt.show()
plt.close()

# ============================================================
# Fast cross-validation statistical evaluation
# ============================================================

print("\nRunning fast cross-validation statistical validation...")
cv_predictions = cross_val_predict(
    clone(final_best_model),
    X_trainval,
    y_trainval,
    cv=cv,
    n_jobs=-1
)

stats_df = pd.DataFrame([
    {'metric': 'accuracy', 'value': accuracy_score(y_trainval, cv_predictions)},
    {'metric': 'macro_precision', 'value': precision_score(y_trainval, cv_predictions, average='macro', zero_division=0)},
    {'metric': 'macro_recall', 'value': recall_score(y_trainval, cv_predictions, average='macro', zero_division=0)},
    {'metric': 'macro_f1', 'value': f1_score(y_trainval, cv_predictions, average='macro', zero_division=0)}
])

print("\nCross-Validation Statistical Results")
display(stats_df)

save_table(
    stats_df,
    TABLE_DIR / 'table_07_fast_statistical_results.csv',
    'Fast statistical validation table'
)

if 'refresh_outputs_summary' in globals():
    refresh_outputs_summary(final=False)

append_summary('Cell 7 completed successfully: learning curve and statistical validation saved.')

print("\n" + "=" * 80)
print("CELL 7 COMPLETED SUCCESSFULLY")
print("=" * 80)

```

## Cell 8 — Global Explainability and SHAP/Feature Importance

```python
# ============================================================
# CELL 8 — GLOBAL EXPLAINABILITY AND SHAP/FEATURE IMPORTANCE
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.inspection import permutation_importance
from sklearn.metrics import make_scorer, f1_score

print("=" * 80)
print("CELL 8 — GLOBAL EXPLAINABILITY AND SHAP/FEATURE IMPORTANCE")
print("=" * 80)

# ============================================================
# Safety fallback utilities
# ============================================================

if 'append_summary' not in globals():
    def append_summary(message):
        print(message)

if 'save_table' not in globals():
    def save_table(df, path, description='table'):
        path.parent.mkdir(parents=True, exist_ok=True)
        df.to_csv(path, index=False)
        print(f"Saved {description}: {path}")

if 'save_figure' not in globals():
    def save_figure(path, description='figure'):
        print(f"Saved {description}: {path}")

# ============================================================
# Recover train-validation data if needed
# ============================================================

if 'X_trainval' not in globals():
    if 'X_train' in globals() and 'X_val' in globals():
        if isinstance(X_train, pd.DataFrame):
            X_trainval = pd.concat([X_train, X_val], axis=0).reset_index(drop=True)
            FEATURE_NAMES = X_trainval.columns.tolist()
        else:
            X_trainval = np.concatenate([np.asarray(X_train), np.asarray(X_val)], axis=0)
            FEATURE_NAMES = [f'feature_{i}' for i in range(X_trainval.shape[1])]
    else:
        raise NameError("X_trainval is missing and X_train/X_val are unavailable. Please run Cells 1–7 first.")

if 'y_trainval' not in globals():
    if 'y_train' in globals() and 'y_val' in globals():
        if isinstance(y_train, pd.Series):
            y_trainval = pd.concat([y_train, y_val], axis=0).values
        else:
            y_trainval = np.concatenate([np.asarray(y_train), np.asarray(y_val)])
    else:
        raise NameError("y_trainval is missing and y_train/y_val are unavailable. Please run Cells 1–7 first.")

required_vars = ['final_best_model', 'FIG_DIR', 'TABLE_DIR']
missing = [v for v in required_vars if v not in globals()]
if missing:
    raise NameError(
        f"Missing required variables for Cell 8: {missing}\n"
        "Please run previous cells first."
    )

# ============================================================
# Recover feature names safely
# ============================================================

print("Recovering feature names...")

if isinstance(X_trainval, pd.DataFrame):
    X_explain = X_trainval.copy()
    feature_names = X_explain.columns.tolist()
else:
    X_explain = pd.DataFrame(X_trainval)
    if 'FEATURE_NAMES' in globals() and len(FEATURE_NAMES) == X_explain.shape[1]:
        feature_names = list(FEATURE_NAMES)
    else:
        feature_names = [f'feature_{i}' for i in range(X_explain.shape[1])]
    X_explain.columns = feature_names

print(f"Number of features: {len(feature_names)}")

# ============================================================
# Compute global feature importance
# ============================================================

importance_values = None
importance_source = None

if hasattr(final_best_model, 'feature_importances_'):
    print("Using native tree-based feature importances...")
    importance_values = final_best_model.feature_importances_
    importance_source = 'native_feature_importances'
elif hasattr(final_best_model, 'coef_'):
    print("Using linear-model coefficients...")
    coef_values = np.abs(final_best_model.coef_)
    importance_values = coef_values.mean(axis=0) if coef_values.ndim > 1 else coef_values
    importance_source = 'linear_coefficients'
else:
    print("Using permutation importance fallback...")
    sample_size = min(5000, len(X_explain))
    X_perm = X_explain.sample(sample_size, random_state=RANDOM_SEED) if hasattr(X_explain, 'sample') else X_explain[:sample_size]
    y_perm = np.asarray(y_trainval)[:sample_size]
    perm_result = permutation_importance(
        final_best_model,
        X_perm,
        y_perm,
        scoring=make_scorer(f1_score, average='macro', zero_division=0),
        n_repeats=5,
        random_state=RANDOM_SEED,
        n_jobs=-1
    )
    importance_values = perm_result.importances_mean
    importance_source = 'permutation_importance_f1_macro'

importance_values = np.asarray(importance_values).reshape(-1)
min_len = min(len(feature_names), len(importance_values))

importance_df = pd.DataFrame({
    'feature': feature_names[:min_len],
    'importance': importance_values[:min_len]
}).sort_values('importance', ascending=False).reset_index(drop=True)

importance_sum = importance_df['importance'].sum()
if importance_sum != 0:
    importance_df['normalized_importance'] = importance_df['importance'] / importance_sum
else:
    importance_df['normalized_importance'] = 0.0
importance_df['rank'] = np.arange(1, len(importance_df) + 1)

print(f"Importance source: {importance_source}")
print("\nTop Features")
display(importance_df.head(20))

save_table(
    importance_df,
    TABLE_DIR / 'table_08_feature_importance.csv',
    'Feature importance table'
)

TOP_K = min(20, len(importance_df))
plot_df = importance_df.head(TOP_K).iloc[::-1]
plt.figure(figsize=(10, 7))
plt.barh(plot_df['feature'], plot_df['normalized_importance'])
plt.xlabel('Normalized Importance')
plt.ylabel('Feature')
plt.title(f'Top {TOP_K} Feature Importances')
plt.tight_layout()
fig_path = FIG_DIR / 'fig_08_feature_importance.png'
plt.savefig(fig_path, dpi=300, bbox_inches='tight')
save_figure(fig_path, 'Feature importance figure')
plt.show()
plt.close()

# ============================================================
# Optional SHAP analysis
# ============================================================

shap_status = 'not_run'
shap_importance_df = None

try:
    import shap
    print("\nRunning optional SHAP explainability...")

    sample_size = min(1000, len(X_explain))
    X_sample = X_explain.sample(sample_size, random_state=RANDOM_SEED)

    try:
        explainer = shap.Explainer(final_best_model, X_sample)
        shap_values = explainer(X_sample)
        vals = np.asarray(shap_values.values)

        if vals.ndim == 3:
            shap_abs = np.mean(np.abs(vals), axis=(0, 2))
        elif vals.ndim == 2:
            shap_abs = np.mean(np.abs(vals), axis=0)
        else:
            shap_abs = np.mean(np.abs(vals), axis=tuple(range(vals.ndim - 1))).reshape(-1)

        min_len = min(len(feature_names), len(shap_abs))
        shap_importance_df = pd.DataFrame({
            'feature': feature_names[:min_len],
            'mean_abs_shap': shap_abs[:min_len]
        }).sort_values('mean_abs_shap', ascending=False).reset_index(drop=True)

        shap_sum = shap_importance_df['mean_abs_shap'].sum()
        if shap_sum != 0:
            shap_importance_df['normalized_mean_abs_shap'] = shap_importance_df['mean_abs_shap'] / shap_sum
        else:
            shap_importance_df['normalized_mean_abs_shap'] = 0.0
        shap_importance_df['rank'] = np.arange(1, len(shap_importance_df) + 1)

        save_table(
            shap_importance_df,
            TABLE_DIR / 'table_08_shap_feature_importance.csv',
            'SHAP feature importance table'
        )

        top_shap = shap_importance_df.head(TOP_K).iloc[::-1]
        plt.figure(figsize=(10, 7))
        plt.barh(top_shap['feature'], top_shap['normalized_mean_abs_shap'])
        plt.xlabel('Normalized Mean |SHAP|')
        plt.ylabel('Feature')
        plt.title(f'Top {TOP_K} SHAP Feature Importances')
        plt.tight_layout()
        shap_fig_path = FIG_DIR / 'fig_08_shap_feature_importance.png'
        plt.savefig(shap_fig_path, dpi=300, bbox_inches='tight')
        save_figure(shap_fig_path, 'SHAP feature importance figure')
        plt.show()
        plt.close()

        shap_status = 'completed'
        print("SHAP analysis completed.")

    except Exception as shap_inner_error:
        shap_status = f'skipped: {shap_inner_error}'
        print(f"SHAP internal failure: {shap_inner_error}")

except Exception as shap_error:
    shap_status = f'skipped: {shap_error}'
    print(f"SHAP skipped: {shap_error}")

if 'save_json' in globals():
    save_json(
        {'importance_source': importance_source, 'shap_status': shap_status},
        OUTPUT_DIR / 'cell_08_explainability_summary.json',
        'Explainability summary'
    )

if 'refresh_outputs_summary' in globals():
    refresh_outputs_summary(final=False)

append_summary('Cell 8 completed successfully: explainability artifacts saved.')

print("\n" + "=" * 80)
print("CELL 8 COMPLETED SUCCESSFULLY")
print("=" * 80)

```

## Cell 9 — Forensic Evidence Logging with Hash Chain

```python
# ============================================================
# CELL 9 — FORENSIC EVIDENCE LOGGING WITH HASH CHAIN
# ============================================================

print("=" * 80)
print("CELL 9 — FORENSIC EVIDENCE LOGGING WITH HASH CHAIN")
print("=" * 80)

X_log = X_test.reset_index(drop=True).copy()
y_log = np.asarray(y_test)
pred_log = np.asarray(y_pred_ton)
if y_proba_ton is not None:
    conf_log = np.max(y_proba_ton, axis=1)
else:
    conf_log = np.ones(len(pred_log))

top_features = importance_df['feature'].head(10).astype(str).tolist()

def stable_hash(obj):
    return hashlib.sha256(json.dumps(obj, sort_keys=True, default=str).encode('utf-8')).hexdigest()

records = []
previous_hash = 'GENESIS'
max_records = min(5000, len(X_log))
for i in range(max_records):
    feature_hash = stable_hash(X_log.iloc[i].to_dict())
    rec = {
        'event_id': str(uuid.uuid4()),
        'case_id': 'FRX_IDS_TON_CASE_001',
        'row_position': int(i),
        'timestamp_utc': utc_now(),
        'model_name': final_best_model_name,
        'true_label_id': int(y_log[i]),
        'true_label_name': str(CLASS_NAMES_TON[int(y_log[i])]),
        'predicted_label_id': int(pred_log[i]),
        'predicted_label_name': str(CLASS_NAMES_TON[int(pred_log[i])]),
        'prediction_correct': bool(y_log[i] == pred_log[i]),
        'confidence': float(conf_log[i]),
        'uncertainty': float(1.0 - conf_log[i]),
        'feature_hash': feature_hash,
        'top_global_features': top_features,
        'previous_record_hash': previous_hash
    }
    rec_hash = stable_hash(rec)
    rec['record_hash'] = rec_hash
    previous_hash = rec_hash
    records.append(rec)

forensic_log_df = pd.DataFrame(records)
log_csv_path = LOG_DIR / 'log_09_forensic_evidence_hash_chain.csv'
log_jsonl_path = LOG_DIR / 'log_09_forensic_evidence_hash_chain.jsonl'
forensic_log_df.to_csv(log_csv_path, index=False)
register_artifact(log_csv_path, 'forensic_log_csv', 'Hash-linked forensic evidence log')
with open(log_jsonl_path, 'w', encoding='utf-8') as f:
    for r in records:
        f.write(json.dumps(r, default=str) + '\n')
register_artifact(log_jsonl_path, 'forensic_log_jsonl', 'Hash-linked forensic evidence JSONL log')

# Verify chain
valid_chain = True
prev = 'GENESIS'
for r in records:
    copy = r.copy()
    saved = copy.pop('record_hash')
    if copy['previous_record_hash'] != prev or stable_hash(copy) != saved:
        valid_chain = False
        break
    prev = saved

save_json({'records': len(records), 'hash_chain_valid': valid_chain, 'final_record_hash': previous_hash}, OUTPUT_DIR / 'cell_09_forensic_log_summary.json', 'Forensic log summary')
refresh_outputs_summary(final=False)
append_summary(f'Cell 9 completed: forensic log saved, hash_chain_valid={valid_chain}.')
display(forensic_log_df.head())
```

## Cell 10 — Forensic Graph and Timeline Artifacts

```python
# ============================================================
# CELL 10 — FORENSIC GRAPH AND TIMELINE ARTIFACTS
# ============================================================

import matplotlib.pyplot as plt

print("=" * 80)
print("CELL 10 — FORENSIC GRAPH AND TIMELINE ARTIFACTS")
print("=" * 80)

CASE_ID = 'FRX_IDS_TON_CASE_001'
graph_df = forensic_log_df.copy()
graph_df['source_host'] = ['src_' + hashlib.sha1(f"{CASE_ID}_src_{i}".encode()).hexdigest()[:8] for i in graph_df['row_position']]
graph_df['destination_host'] = ['dst_' + hashlib.sha1(f"{CASE_ID}_dst_{i}".encode()).hexdigest()[:8] for i in graph_df['row_position']]
graph_df['attack_node'] = 'attack_' + graph_df['predicted_label_name'].astype(str)

nodes = []
edges = []
for _, r in graph_df.iterrows():
    nodes.extend([
        {'node_id': r['source_host'], 'node_type': 'source_host'},
        {'node_id': r['destination_host'], 'node_type': 'destination_host'},
        {'node_id': r['attack_node'], 'node_type': 'predicted_attack_class'},
        {'node_id': 'evidence_' + r['event_id'], 'node_type': 'evidence_record'},
    ])
    edges.extend([
        {'source': r['source_host'], 'target': r['destination_host'], 'edge_type': 'communicates_with', 'event_id': r['event_id'], 'confidence': r['confidence']},
        {'source': r['destination_host'], 'target': r['attack_node'], 'edge_type': 'associated_with_attack', 'event_id': r['event_id'], 'confidence': r['confidence']},
        {'source': r['attack_node'], 'target': 'evidence_' + r['event_id'], 'edge_type': 'supported_by_evidence', 'event_id': r['event_id'], 'confidence': r['confidence']},
    ])

node_df = pd.DataFrame(nodes).drop_duplicates()
edge_df = pd.DataFrame(edges)
save_table(node_df, TABLE_DIR / 'table_10_graph_nodes.csv', 'Forensic graph nodes')
save_table(edge_df, TABLE_DIR / 'table_10_graph_edges.csv', 'Forensic graph edges')

attack_summary_df = graph_df.groupby('predicted_label_name').agg(
    evidence_records=('event_id', 'count'),
    mean_confidence=('confidence', 'mean'),
    incorrect_records=('prediction_correct', lambda s: int((~s.astype(bool)).sum()))
).reset_index().sort_values('evidence_records', ascending=False)
save_table(attack_summary_df, TABLE_DIR / 'table_10_attack_summary.csv', 'Forensic attack summary')

plt.figure(figsize=(10, 6))
plot_df = attack_summary_df.sort_values('evidence_records', ascending=True)
plt.barh(plot_df['predicted_label_name'], plot_df['evidence_records'])
plt.xlabel('Evidence records')
plt.ylabel('Predicted attack class')
plt.title('Forensic Evidence Records by Predicted Attack Class')
plt.tight_layout()
fig_graph_path = FIG_DIR / 'fig_10_forensic_attack_summary.png'
plt.savefig(fig_graph_path, dpi=300, bbox_inches='tight')
plt.show()
save_figure(fig_graph_path, 'Forensic attack summary figure')

# Timeline: use generated timestamps and order
timeline_df = graph_df[['event_id', 'timestamp_utc', 'predicted_label_name', 'confidence', 'record_hash']].copy()
timeline_df['timestamp_utc'] = pd.to_datetime(timeline_df['timestamp_utc'], utc=True, errors='coerce')
timeline_df = timeline_df.dropna(subset=['timestamp_utc']).sort_values('timestamp_utc').reset_index(drop=True)
timeline_df['timeline_order'] = np.arange(len(timeline_df))
timeline_df['minute_bucket'] = timeline_df['timestamp_utc'].dt.floor('min')
save_table(timeline_df, TABLE_DIR / 'table_10_timeline_events.csv', 'Forensic timeline events')

timeline_agg = timeline_df.groupby(['minute_bucket', 'predicted_label_name']).size().reset_index(name='evidence_records')
save_table(timeline_agg, TABLE_DIR / 'table_10_timeline_aggregation.csv', 'Timeline aggregation by minute and class')

plt.figure(figsize=(10, 5))
density = timeline_df.groupby('minute_bucket').size().reset_index(name='evidence_records')
plt.plot(density['minute_bucket'], density['evidence_records'], marker='o')
plt.xlabel('Time')
plt.ylabel('Evidence records')
plt.title('Forensic Timeline Evidence Density')
plt.xticks(rotation=30)
plt.tight_layout()
fig_timeline_path = FIG_DIR / 'fig_10_timeline_density.png'
plt.savefig(fig_timeline_path, dpi=300, bbox_inches='tight')
plt.show()
save_figure(fig_timeline_path, 'Forensic timeline density figure')

save_json({'case_id': CASE_ID, 'nodes': len(node_df), 'edges': len(edge_df), 'timeline_events': len(timeline_df)}, OUTPUT_DIR / 'cell_10_graph_timeline_summary.json', 'Graph and timeline summary')
refresh_outputs_summary(final=False)
append_summary('Cell 10 completed: forensic graph and timeline artifacts saved.')
display(attack_summary_df)
```

## Cell 11 — UNSW-NB15 External Validation Preparation

```python
# ============================================================
# CELL 11 — UNSW-NB15 EXTERNAL VALIDATION PREPARATION
# ============================================================

from sklearn.preprocessing import LabelEncoder

print("=" * 80)
print("CELL 11 — UNSW-NB15 EXTERNAL VALIDATION PREPARATION")
print("=" * 80)

assert 'attack_cat' in unsw_train_df.columns, "UNSW column 'attack_cat' not found."
assert 'attack_cat' in unsw_test_df.columns, "UNSW column 'attack_cat' not found."

unsw_train = unsw_train_df.copy().replace([np.inf, -np.inf], np.nan)
unsw_test = unsw_test_df.copy().replace([np.inf, -np.inf], np.nan)

# Remove records with missing target, then align columns
unsw_train = unsw_train.dropna(subset=['attack_cat']).copy()
unsw_test = unsw_test.dropna(subset=['attack_cat']).copy()

UNSW_TARGET = 'attack_cat'
unsw_drop_cols = [c for c in ['attack_cat', 'label', 'id'] if c in unsw_train.columns]
common_features = [c for c in unsw_train.columns if c in unsw_test.columns and c not in unsw_drop_cols]

X_unsw_train_raw = unsw_train[common_features].copy()
X_unsw_test_raw = unsw_test[common_features].copy()

le_unsw = LabelEncoder()
y_unsw_train = le_unsw.fit_transform(unsw_train[UNSW_TARGET].astype(str))
# Keep only test rows with target classes present in training
known_classes = set(le_unsw.classes_)
mask_known = unsw_test[UNSW_TARGET].astype(str).isin(known_classes).values
X_unsw_test_raw = X_unsw_test_raw.loc[mask_known].copy()
y_unsw_test = le_unsw.transform(unsw_test.loc[mask_known, UNSW_TARGET].astype(str))
CLASS_NAMES_UNSW = le_unsw.classes_

# Encoding based on UNSW training only
cat_unsw = X_unsw_train_raw.select_dtypes(exclude=[np.number]).columns.tolist()
unsw_feature_encoders = {}
X_unsw_train = X_unsw_train_raw.copy()
X_unsw_test = X_unsw_test_raw.copy()
for c in cat_unsw:
    le = LabelEncoder()
    X_unsw_train[c] = X_unsw_train[c].astype(str)
    X_unsw_test[c] = X_unsw_test[c].astype(str)
    le.fit(X_unsw_train[c])
    classes = set(le.classes_)
    X_unsw_test[c] = X_unsw_test[c].apply(lambda x: x if x in classes else 'UNKNOWN')
    if 'UNKNOWN' not in le.classes_:
        le.classes_ = np.append(le.classes_, 'UNKNOWN')
    X_unsw_train[c] = le.transform(X_unsw_train[c])
    X_unsw_test[c] = le.transform(X_unsw_test[c])
    unsw_feature_encoders[c] = le

for Xp in [X_unsw_train, X_unsw_test]:
    for c in Xp.columns:
        Xp[c] = pd.to_numeric(Xp[c], errors='coerce')

unsw_medians = X_unsw_train.median(numeric_only=True).fillna(0)
X_unsw_train = X_unsw_train.fillna(unsw_medians).fillna(0).astype(np.float32)
X_unsw_test = X_unsw_test.fillna(unsw_medians).fillna(0).astype(np.float32)

unsw_class_df = pd.DataFrame({
    'class_id': np.arange(len(CLASS_NAMES_UNSW)),
    'class_name': CLASS_NAMES_UNSW,
    'train_count': [(y_unsw_train == i).sum() for i in range(len(CLASS_NAMES_UNSW))],
    'test_count': [(y_unsw_test == i).sum() for i in range(len(CLASS_NAMES_UNSW))]
})
save_table(unsw_class_df, TABLE_DIR / 'table_11_unsw_class_distribution.csv', 'UNSW-NB15 class distribution')
save_json({'train_shape': X_unsw_train.shape, 'test_shape': X_unsw_test.shape, 'classes': CLASS_NAMES_UNSW.tolist(), 'features': len(common_features)}, OUTPUT_DIR / 'cell_11_unsw_preparation_summary.json', 'UNSW preparation summary')
refresh_outputs_summary(final=False)
append_summary('Cell 11 completed: UNSW-NB15 external validation data prepared.')
display(unsw_class_df)
```

## Cell 12 — UNSW-NB15 External Validation Training and Evaluation

```python
# ============================================================
# CELL 12 — UNSW-NB15 EXTERNAL VALIDATION TRAINING AND EVALUATION
# ============================================================

import joblib
from sklearn.metrics import classification_report, confusion_matrix

print("=" * 80)
print("CELL 12 — UNSW-NB15 EXTERNAL VALIDATION TRAINING AND EVALUATION")
print("=" * 80)

# Model for concise external benchmark validation
try:
    from lightgbm import LGBMClassifier
    unsw_model_name = 'LightGBM_UNSW'
    unsw_model = LGBMClassifier(
        objective='multiclass', num_class=len(CLASS_NAMES_UNSW), n_estimators=300,
        learning_rate=0.05, num_leaves=63, subsample=0.9, colsample_bytree=0.9,
        random_state=RANDOM_SEED, n_jobs=-1, verbose=-1
    )
except Exception:
    unsw_model_name = 'ExtraTrees_UNSW'
    unsw_model = ExtraTreesClassifier(n_estimators=300, random_state=RANDOM_SEED, n_jobs=-1, class_weight='balanced')

append_summary(f'Cell 12 UNSW training started: {unsw_model_name}')
t0 = time.perf_counter()
unsw_model.fit(X_unsw_train, y_unsw_train)
unsw_train_time = time.perf_counter() - t0

y_unsw_pred = unsw_model.predict(X_unsw_test)
if hasattr(unsw_model, 'predict_proba'):
    y_unsw_proba = unsw_model.predict_proba(X_unsw_test)
else:
    y_unsw_proba = None

unsw_metrics = {
    'dataset': 'UNSW-NB15',
    'model': unsw_model_name,
    'train_time_seconds': unsw_train_time,
    'accuracy': accuracy_score(y_unsw_test, y_unsw_pred),
    'balanced_accuracy': balanced_accuracy_score(y_unsw_test, y_unsw_pred),
    'precision_macro': precision_score(y_unsw_test, y_unsw_pred, average='macro', zero_division=0),
    'recall_macro': recall_score(y_unsw_test, y_unsw_pred, average='macro', zero_division=0),
    'f1_macro': f1_score(y_unsw_test, y_unsw_pred, average='macro', zero_division=0),
    'f1_weighted': f1_score(y_unsw_test, y_unsw_pred, average='weighted', zero_division=0)
}
if y_unsw_proba is not None:
    try:
        unsw_metrics['roc_auc_ovr_weighted'] = roc_auc_score(y_unsw_test, y_unsw_proba, multi_class='ovr', average='weighted')
    except Exception:
        unsw_metrics['roc_auc_ovr_weighted'] = None

unsw_metrics_df = pd.DataFrame([unsw_metrics])
save_table(unsw_metrics_df, TABLE_DIR / 'table_12_unsw_external_validation_metrics.csv', 'UNSW external validation metrics')

unsw_report_df = pd.DataFrame(classification_report(y_unsw_test, y_unsw_pred, target_names=CLASS_NAMES_UNSW, output_dict=True, zero_division=0)).T
save_table(unsw_report_df.reset_index().rename(columns={'index':'class'}), TABLE_DIR / 'table_12_unsw_classification_report.csv', 'UNSW classification report')

unsw_cm = confusion_matrix(y_unsw_test, y_unsw_pred)
unsw_cm_df = pd.DataFrame(unsw_cm, index=CLASS_NAMES_UNSW, columns=CLASS_NAMES_UNSW)
save_table(unsw_cm_df.reset_index().rename(columns={'index':'true_class'}), TABLE_DIR / 'table_12_unsw_confusion_matrix.csv', 'UNSW confusion matrix')

plt.figure(figsize=(9, 7))
plt.imshow(unsw_cm, interpolation='nearest')
plt.title(f'UNSW-NB15 Confusion Matrix — {unsw_model_name}')
plt.xlabel('Predicted class')
plt.ylabel('True class')
plt.xticks(range(len(CLASS_NAMES_UNSW)), CLASS_NAMES_UNSW, rotation=45, ha='right')
plt.yticks(range(len(CLASS_NAMES_UNSW)), CLASS_NAMES_UNSW)
plt.colorbar()
plt.tight_layout()
fig_unsw_cm = FIG_DIR / 'fig_12_unsw_confusion_matrix.png'
plt.savefig(fig_unsw_cm, dpi=300, bbox_inches='tight')
plt.show()
save_figure(fig_unsw_cm, 'UNSW-NB15 confusion matrix figure')

unsw_model_path = MODEL_DIR / f'model_12_{unsw_model_name}.joblib'
joblib.dump(unsw_model, unsw_model_path)
register_artifact(unsw_model_path, 'model_joblib', 'UNSW external validation model')
save_json(unsw_metrics, OUTPUT_DIR / 'cell_12_unsw_external_validation_metrics.json', 'UNSW external validation metrics JSON')
refresh_outputs_summary(final=False)
append_summary('Cell 12 completed: UNSW-NB15 external validation results saved.')
display(unsw_metrics_df)
```
